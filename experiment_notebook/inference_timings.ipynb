```python
# cnn network inference timings

timings = [1, 10, 100, 1000]
sum_timings_cnn = [
    0.186144,
    1.883154,
    22.098292,
    198.732134,
]
sum_timings_cnn_stat = [
    0.200398,
    1.997745,
    21.725904,
    219.862069,
]

sum_timings_cnn_lstm = [
    0.166968,
    1.667871,
    16.722465,
    174.231837,
]

average_inference_cnn = [
    0.186144,
    0.188315,
    0.220983,
    0.198732,
]

average_inference_cnn_stat = [
    0.200398,
    0.199774,
    0.217259,
    0.219862,
]

average_inference_cnn_lstm = [
    0.166968,
    0.166787,
    0.167225,
    0.174232,
]

figure, (ax1, ax2) = plt.subplots(1, 2)
figure.set_figheight(4)
figure.set_figwidth(12)
figure.set_dpi(300)
figure.tight_layout(pad=2.0)

x = np.arange(len(timings))
ax1.set_xlabel('Number of inferences')
ax1.set_ylabel('Time in seconds (s)')
width = 0.30
rect1 = ax1.bar(x - 0.30, sum_timings_cnn, width=width, label="CNN")
rect2 = ax1.bar(x, sum_timings_cnn_stat, width=width, label="CNN-Statistics")
rect3 = ax1.bar(x + 0.30, sum_timings_cnn_stat,
                width=width, label="CNN-LSTM")
ax1.set_xticks(x, timings)

ax1.grid(linestyle="--", linewidth=0.25)
ax1.legend(loc=2, prop={"size": 10})
ax1.set_title('Time taken vs Number of inferences in RaspberryPi 3')

ax2.plot(timings, average_inference_cnn, label="CNN", linestyle="solid")
ax2.plot(timings, average_inference_cnn_stat, label="CNN-Statistics", linestyle="dashed")
ax2.plot(timings, average_inference_cnn_lstm, label="CNN-LSTM", linestyle="dashdot")
ax2.set_xlabel('Number of inferences')
ax2.set_ylabel('Time in seconds (s)')

ax2.grid('linestyle="--', linewidth=0.25)
ax2.legend(loc=1, prop={"size": 10})
ax2.set_title('Average time taken vs Number of inferences in RaspberryPi 3')



plt.show()
```
